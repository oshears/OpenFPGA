```python
import json

fh = open("/home/oshears/Documents/openfpga/OpenFPGA/debug/bitstream_validator/results/tracedPaths.json","r")

json_string = fh.read()

data = json.loads(json_string)
```

```python
data["nodes"][0]
```

```python
data["nodes"][0]
N = len(data["nodes"])
Edges = data["links"]
```

```python
text = [data["nodes"][k]["name"] for k in range(N)]
```

```python
Xn=[data["nodes"][k]["output"] for k in range(N)]# x-coordinates of nodes
Yn=[data["nodes"][k]["level"] for k in range(N)]# y-coordinates
Zn=[data["nodes"][k]["path_direction"] for k in range(N)]# z-coordinates
Xe=[]
Ye=[]
Ze=[]
for e in Edges:
    Xe+=[data["nodes"][e[0]]["output"], data["nodes"][e[1]]["output"], None]# x-coordinates of edge ends
    Ye+=[data["nodes"][e[0]]["level"], data["nodes"][e[1]]["level"], None]
    Ze+=[data["nodes"][e[0]]["path_direction"], data["nodes"][e[1]]["path_direction"], None]
```

```python
from plotly.offline import iplot
import plotly.graph_objs as go

trace1=go.Scatter3d(
               x=Xe,
               y=Ye,
               z=Ze,
               mode='lines',
               line=dict(color='rgb(125,125,125)', width=5),
               hoverinfo='none'
               )

trace2 = go.Scatter3d(
    x=Xn, 
    y=Yn, 
    z=Zn,
    text=text,
    hoverinfo='text',
    mode='markers',
    name="Points",
    marker=dict(symbol='circle',
                             size=6,
                             color=Yn,
                             colorscale='Viridis',
                             line=dict(color='rgb(50,50,50)', 
                            width=0.5)
                             ),
    )

axis=dict(showbackground=False,
          showline=False,
          zeroline=False,
          showgrid=False,
          showticklabels=False,
          title=''
          )

layout = go.Layout(
         title="3D Visualization of Routing Paths",
         width=1000,
         height=1000,
         showlegend=False,
         scene=dict(
             xaxis=dict(axis),
             yaxis=dict(axis),
             zaxis=dict(axis),
        ),
     margin=dict(
        t=100
    ),
    hovermode='closest',
    )


graph_data = [trace1, trace2]

fig = go.Figure(graph_data, layout=layout)

iplot(fig)
```
